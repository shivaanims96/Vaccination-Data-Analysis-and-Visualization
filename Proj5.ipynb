```python
import mysql.connector
from io import StringIO
```

```python
def create_connection(db_user, db_password, db_host, db_name):
    try:
        connection = mysql.connector.connect(
            user=db_user,
            password=db_password,
            host=db_host,
            database=db_name,
            auth_plugin='mysql_native_password',
            use_pure=True
        )
        if connection.is_connected():
            print("Connected to MySQL database")
            return connection
        else:
            print("Failed to connect to MySQL database")
            return None
    except mysql.connector.Error as error:
        print(f"Error: {error}")
        return None
```

```python
def pandas_to_mysql(connection, table_name, csv_data):
    """
    Inserts data from a CSV string into a MySQL table.

    Args:
        connection: MySQL connection object.
        table_name (str): The name of the table to insert into.
        csv_data (str):  A string containing the CSV data.
    """

    df = pd.read_csv(StringIO(csv_data), sep='\t')  # Read CSV data into Pandas DataFrame
    cursor = connection.cursor()

    try:
        cols = ",".join([f"`{col}`" for col in df.columns])

        for _, row in df.iterrows():
            values = tuple(row.fillna(None)) # Replace NaN with None
            placeholders = ', '.join(['%s'] * len(values))
            insert_query = f"INSERT INTO {table_name} ({cols}) VALUES ({placeholders})"
            cursor.execute(insert_query, values)

        connection.commit()
        print(f"Data successfully inserted into table '{table_name}'")

    except mysql.connector.Error as error:
        print(f"Error: {error}")
        connection.rollback()  # Rollback changes in case of error

    finally:
        cursor.close()
```

```python
def create_table_func(create_table_arg,db):
     try :
          conn = sqlite3.connect(db)
          cursor = conn.cursor()
          cursor.execute(create_table_arg)
          conn.commit()
     except sqlite3.Error as e:
          return f"An error occurred: {e}"
          conn.rollback()
     finally:
          cursor.close()
          conn.close()
     
     return "Table Created successfully."
```

```python
def insert_sql(table_name, db, dfs):
    try :
          conn = sqlite3.connect(db)
          cursor = conn.cursor()
          dfs.to_sql(table_name,conn,if_exists='append',index=False)
          conn.commit()
    except sqlite3.Error as e:
        return f"An error occurred: {e}"
        conn.rollback()
    finally:
        cursor.close()
        conn.close()
     
    return "Records Inserted successfully."
```
